```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

data_path = '/kaggle/input/harry-potter-books'
files = os.listdir(data_path)
print(files)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in files:
    
    f = open(os.path.join(data_path,filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    # tokens should be a list of words
    return [word for word in tokens if word not in stop_words]
```

```python
filtered_story = [remove_stopwords(sent) for sent in story]
```

```python
filtered_story
```

```python
model = gensim.models.Word2Vec(
    window = 5,
    min_count = 2
)
```

```python
model.build_vocab(filtered_story)  # Extracted unique words from corpus
```

```python
model.train(filtered_story,total_examples=model.corpus_count , epochs =100)
```

```python
model.wv.most_similar('harry')
```
